```python
import numpy as np
import pandas as pd
```

```python
cities_path = 'restaurant_cities.csv'
establishments_path = 'restaurant_establishments.csv'
businesses_path = 'yelp_academic_dataset_business.json'
reviews_path = 'yelp_review.csv'
```

```python
cities = pd.read_csv(cities_path, sep=';', index_col = 0)
cities.head()
```

```python
cities.info()
```

```python
establishments = pd.read_csv(establishments_path, sep=';', index_col=0)
establishments.rename(columns={'restaurant_name':'name'}, inplace=True)
establishments.head()
```

```python
establishments.info()
```

```python
businesses = pd.read_json(businesses_path, lines=True)
businesses.head()
```

```python
#Create new df that contains key data for merge with inspection data later
business_tomerge = businesses[['business_id', 'name', 'address', 'city']]
business_tomerge.head()
```

```python
businesses.info()
```

```python
reviews = pd.read_csv(reviews_path)
reviews.info()
```

```python
#Create new df that contains only review id, business id and text to merge with inspection data
reviews_tomerge = reviews[['review_id', 'business_id', 'text']]
reviews_tomerge.head()
```

```python
reviews.business_id.value_counts()
```

```python
review_locs = pd.DataFrame()
business_locs = pd.DataFrame()
review_locs['business_id'] = reviews['business_id']
business_locs['business_id'] = businesses['business_id']
business_locs['city'] = businesses['city']
```

```python
business_locs.business_id.value_counts()
```

```python
merged = pd.merge(review_locs, business_locs, on='business_id')
```

```python
merged.info()
```

```python
merged.city.value_counts()
```

```python
dropped_df = pd.merge(review_locs, business_locs,on='business_id', indicator=True, how='outer')
```

```python
#Merge of the two '_tomerge' sets
Yelp_tomerge = pd.merge(reviews_tomerge, business_tomerge, on='business_id')
Yelp_tomerge.info()
```

```python
#Merge Yelp_tomerge with establishments
df = establishments.merge(Yelp_tomerge, how='inner', left_on=['name', 'address', 'city_name'], right_on=['name', 'address', 'city'])
df.drop_duplicates(inplace=True)
df.info()
#df.dropna(how='all')
```
